```python
from bs4 import BeautifulSoup
import urllib.request as req
url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
res =req.urlopen(url)
bs=BeautifulSoup(res, 'html.parser')
title=bs.find('title').text
wf = bs.find('wf').text
print(title)
print(wf)
```

```python
result1=[]
result2=[]
result3=[]
result4=[]
result5=[]
# 'data' 라는 태그를 모두 찾기
#<data> 
# <tmef>0</tmef>
# <wf>1</wf>
#</data>
items=bs.find_all('data')
for item in items:
    # 'tmef' 라는 태그를 모두 찾기
    times=item.find_all('tmef')
    wfs = item.find_all('wf')
    tmns = item.find_all('tmn')
    tmxs = item.find_all('tmx')
    rnsts = item.find_all('rnst')
    for idx in range(len(times)):
        result1.append(times[idx].text)
        result2.append(wfs[idx].text)
        result3.append(tmns[idx].text)
        result4.append(tmxs[idx].text)
        result5.append(rnsts[idx].text)
# zip : 같은 인덱스끼리 결합
result = zip(result1, result2, result3, result4, result5)
rows=list(result)
rows
```

```python
import pandas as pd
df = pd.DataFrame(rows, columns=['일시', '날씨', '최저', '최고', '강수확률'])
df.head()
```

```python
# csv로 저장
df.to_csv('c:/data/weather/data.csv', encoding='ms949')
```

```python
import urllib.request as req
import urllib.request
from urllib.request import urlopen
import time

codes=[]
for num in range(51, 53):
    url='https://www.auction.co.kr/category/category{}.html'.format(num)
    print(url)
    res=req.urlopen(url)
    bs=BeautifulSoup(res, 'html.parser', from_encoding='ms949')
    category = bs.select('div.layer-body > div > div > div > ul > li > ul > li > a')
    time.sleep(1)
    for cate in category:
        name = cate.text
        href = cate.attrs['href'].split('=')[1]
        codes.append(href)
        print('상품분류명:' +name)
        print('분류코드:' +href)
    
    for cate in codes:
        html = urlopen('http://browse.auction.co.kr/list?category='+cate)
        bs = BeautifulSoup(html.read(), 'html.parser')     
        tags = bs.find_all('img')
        for img in tags:
            if 'http' not in img['src']:
                src = 'http:' + img['src']
            else:
                src=img['src']
            name = src.rsplit('/', 1)[-1]
            arr = name.split('?')
            try:
                urllib.request.urlretrieve(src, 'c:/data/images/' + arr[0])
            except:
                print(name)
                pass
```
