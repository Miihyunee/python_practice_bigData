```python
import pandas as pd
from sklearn.datasets import fetch_openml

# https://www.openml.org/d/40945
X, y = fetch_openml("titanic", version=1, as_frame=True, return_X_y=True)
df=pd.concat([X,y],axis=1)
df
```

```python
df.columns
```

```python
train_cols = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']
X = df[train_cols] # 독립변수
y = df["survived"]
y.value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
colors = ["yellow", "red"]

sns.countplot(x='survived', data=df, palette=colors)
```

```python
# 언더샘플링
from imblearn.under_sampling import RandomUnderSampler


X_sample, y_sample = RandomUnderSampler(random_state=0).fit_resample(X, y)
X_samp = pd.DataFrame(data=X_sample,columns=train_cols )
y_samp = pd.DataFrame(data=y_sample,columns=['survived'])
df2=pd.concat([X_samp,y_samp],axis=1)
df2.survived.value_counts()
```

```python
X = df2[train_cols] # 독립변수
y = df2["survived"]
```

```python
colors = ["yellow", "red"]
sns.countplot(x='survived', data=df2, palette=colors)
```

```python
X = X_samp[train_cols] # 독립변수
y = y_samp["survived"]
```

```python
# 데이터 전처리 자동화
# 연속된 변환을 순서대로 처리하는 클래스
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer

# 결측값 처리 - 나이, 요금(중위수로 처리)
# 스케일링
numeric_fratures = ['age', 'fare']
numeric_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])

# 범주형 변수 - 원핫인코딩
# embarked ; {'C', 'S', 'Q'}
# sex : {'female', 'male'}
# pclass ; {1, 2, 3}
# handle_unknown = 'ignore' 변환 중에 에러 발생할 경우 무시하고 계속 진행
categorical_features = ['embarked', 'sex', 'pclass']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_fratures),
        ('cat', categorical_transformer, categorical_features)])

# 전처리 후 로지스틱 회귀분석 분류모형
clf = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', LogisticRegression())])
clf
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)

clf.fit(X_train, y_train)

print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
# 교차검증과 모형의 튜닝
# 하이퍼 파라미터 최적화
from sklearn.model_selection import GridSearchCV

# 숫자 결측값 처리 
# Cost
param_grid = {
    'preprocessor__num__imputer__strategy' : ['mean', 'median'],
    'classifier__C' : [0.0001, 0.001, 0.01, 0.1, 1.0, 10, 100],
}

grid_search = GridSearchCV(clf, param_grid, cv=10)
grid_search.fit(X, y)
grid_search.cv_results_["params"]
```

```python
scores = grid_search.cv_results_["mean_test_score"]
scores
```

```python
print(max(scores))
plt.plot(scores)
```

```python
print(grid_search.best_score_)
print(grid_search.best_params_)
```
